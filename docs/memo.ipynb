## Implicit scheme

\begin{equation}
  f_{i,j}^{k+1}
  = f_{i,j}^{k}
  - \frac{\Delta t}{\Delta x}\, v_{j}\,
    \bigl(
      \hat{f}_{\,i+\tfrac12,j}^{\,k+1}
      -
      \hat{f}_{\,i-\tfrac12,j}^{\,k+1}
    \bigr)
  + \frac{\Delta t}{\tau_i^{k+1}}\,
    \bigl(
      f_{M,i,j}^{k+1}
      -
      f_{i,j}^{k+1}
    \bigr),
  \qquad
  i = 1,\dots,N_x-2,\; j = 0,\dots,N_v-1.
\end{equation}

\begin{equation}
\hat{f}_{\,i+\tfrac12,j}^{\,k+1}
=
\begin{cases}
  f_{\,i,j}^{\,k+1}, & v_j > 0,\\[6pt]
  f_{\,i+1,j}^{\,k+1}, & v_j < 0,
\end{cases}
\qquad
\hat{f}_{\,i-\tfrac12,j}^{\,k+1}
=
\begin{cases}
  f_{\,i-1,j}^{\,k+1}, & v_j > 0,\\[6pt]
  f_{\,i,j}^{\,k+1}, & v_j < 0.
\end{cases}
\end{equation}

\begin{equation}
-\,\beta_j\,f_{\,i+1,j}^{k+1}
+\Bigl(1+\alpha_j+\beta_j+\tfrac{\Delta t}{\tau_i^{k+1}}\Bigr)\,
   f_{\,i,j}^{k+1}
-\,\alpha_j\,f_{\,i-1,j}^{k+1}
=
f_{\,i,j}^{k}
+\frac{\Delta t}{\tau_i^{k+1}}\,
  f_{M,i,j}^{k+1}
\end{equation}

\begin{equation}
\alpha_j=\frac{\Delta t}{\Delta x}\,\max(v_j,0),
\qquad
\beta_j =\frac{\Delta t}{\Delta x}\,\max(-v_j,0).
\end{equation}

\begin{equation}
  \alpha_j \;=\; \frac{\Delta t}{\Delta x}\,\max\!\bigl(v_j,0\bigr),
  \qquad
  \beta_j  \;=\; \frac{\Delta t}{\Delta x}\,\max\!\bigl(-v_j,0\bigr)
\end{equation}

\begin{equation}
\begin{aligned}
  a_{i,j}^{(z)} &= \beta_j,
  &\quad&(i=0,\dots,N_x-2) &\text{(上対角)},\\[4pt]
  b_{i,j}^{(z)} &= 1+\alpha_j+\beta_j+\dfrac{\Delta t}{\tau_i^{(z)}},
  &\quad&(i=0,\dots,N_x-1) &\text{(主対角)},\\[4pt]
  c_{i,j}^{(z)} &= \alpha_j,
  &\quad&(i=1,\dots,N_x-1) &\text{(下対角)}
\end{aligned}
\end{equation}

\begin{equation}
  \mathcal{A}_j^{(z)}
  \;=\;
  \begin{bmatrix}
    b_{0,j}^{(z)}     & -a_{0,j}^{(z)}    &                     &                     \\
    -c_{1,j}^{(z)}    &  b_{1,j}^{(z)}    & -a_{1,j}^{(z)}      &                     \\
                      &  \ddots           &  \ddots             & \ddots              \\
                      &                   & -c_{N_x-2,j}^{(z)}  & b_{N_x-1,j}^{(z)}
  \end{bmatrix}_{N_x\times N_x}
\end{equation}

 変数ベクトル（未知量）
\begin{equation}
  \mathbf{f}^{(z+1)}_{j}
  \;=\;
  \begin{bmatrix}
    f^{(z+1)}_{0,j}\\
    f^{(z+1)}_{1,j}\\
    \vdots\\
    f^{(z+1)}_{N_x-1,j}
  \end{bmatrix},
  \qquad
  j = 0,\dots,N_v-1 .
\end{equation}

 右辺（ソース）ベクトル
\begin{equation}
  \mathbf{d}^{(z)}_{j}
  \;=\;
  \begin{bmatrix}
    d^{(z)}_{0,j}\\
    d^{(z)}_{1,j}\\
    \vdots\\
    d^{(z)}_{N_x-1,j}
  \end{bmatrix},
  \qquad
  d^{(z)}_{i,j}
  \;=\;
  f^{k}_{i,j}
  +\frac{\Delta t}{\tau^{(z)}_{i}}\,
   f_{M,i,j}^{(z)} .
\end{equation}

## HOLO

### 差分式(HO)

\begin{equation}
\frac{f_{i,j}^{\,k+1,z}-f_{i,j}^{\,k,z}}{\Delta t}
+\frac{v_j}{2\Delta x}\!\left(
  \hat f_{\,i+\tfrac12,j}^{\,k+1,z}-\hat f_{\,i-\tfrac12,j}^{\,k+1,z}
 +\hat f_{\,i+\tfrac12,j}^{\,k}-\hat f_{\,i-\tfrac12,j}^{\,k}
\right)
=
\theta\,\frac{f_{M,i,j}^{LO\,k+1,z-1}-f_{i,j}^{\,k+1,z}}{\tau_i^{LO\,k+1,z-1}}
+(1-\theta)\,\frac{f_{M,i,j}^{HO\,k}-f_{i,j}^{\,k}}{\tau_i^{HO\,k}}
\end{equation}

\begin{equation}
\quad
i=1,\dots,N_x-2,\ j=0,\dots,N_v-1 .
\end{equation}

\begin{equation}
\mathrm{EXP}_{i,j}
= f_{i,j}^{\,k,z}
 - \frac{v_j \,\Delta t}{2\,\Delta x}
   \Bigl( \hat f_{\,i+\tfrac12,j}^{\,k} - \hat f_{\,i-\tfrac12,j}^{\,k} \Bigr)
 + (1-\theta)\,\frac{\Delta t}{\tau_i^{\,HO,k}}
   \Bigl( f_{M,i,j}^{\,HO,k} - f_{i,j}^{\,k} \Bigr).
\end{equation}


\begin{equation}
\hat f_{\,i+\tfrac12,j}^{\,k+1,z}
=
\begin{cases}
f_{i,j}^{\,k+1,z},   & v_j > 0,\\[2mm]
f_{i+1,j}^{\,k+1,z}, & v_j < 0,
\end{cases}
\qquad
\hat f_{\,i-\tfrac12,j}^{\,k+1,z}
=
\begin{cases}
f_{i-1,j}^{\,k+1,z}, & v_j > 0,\\[2mm]
f_{i,j}^{\,k+1,z},   & v_j < 0.
\end{cases}
\end{equation}


\begin{equation}
\alpha_j = \frac{\Delta t}{2\,\Delta x}\,\max(v_j,0),
\qquad
\beta_j  = \frac{\Delta t}{2\,\Delta x}\,\max(-v_j,0).
\end{equation}


\begin{equation}
-\beta_{j}\,f_{i+1,j}^{\,k+1,z}
+\Bigl(1+\alpha_{j}+\beta_{j}
      +\theta^{\,HO,k}\,\frac{\Delta t}{\tau_{i}^{\,LO,k+1,z-1}}\Bigr) f_{i,j}^{\,k+1,z}
-\alpha_{j}\,f_{i-1,j}^{\,k+1,z}
=
\mathrm{EXP}_{i,j}
+\theta^{\,HO,k}\,\frac{\Delta t}{\tau_{i}^{\,LO,k+1,z-1}}\,
  f_{M,i,j}^{\,LO,k+1,z-1}\,.
\end{equation}


\begin{equation}
\begin{aligned}
  a_{i,j}^{(k+1,z)} &= \beta_j,
  &\quad&(i=0,\dots,N_x-2) &\text{(上対角)},\\[4pt]
  b_{i,j}^{(k+1,z)} &= 1+\alpha_{j}+\beta_{j}
      +\theta^{\,HO,k}\,\frac{\Delta t}{\tau_{i}^{\,LO,k+1,z-1}} &\quad&(i=0,\dots,N_x-1) &\text{(主対角)},\\[4pt]
  c_{i,j}^{(k+1,z)} &= \alpha_j,
  &\quad&(i=1,\dots,N_x-2) &\text{(下対角)}
\end{aligned}
\end{equation}

\begin{equation}
\mathcal{A}_{j}^{(k+1,z)}
=
\begin{bmatrix}
 b_{1,j}^{(k+1,z)}   & -a_{1,j}^{(k+1,z)} &                     &                     \\
 -c_{2,j}^{(k+1,z)}  &  b_{2,j}^{(k+1,z)} & -a_{2,j}^{(k+1,z)}  &                     \\
                     &  \ddots           &  \ddots             & \ddots              \\
                     &                   & -c_{N_x-2,j}^{(k+1,z)} & b_{N_x-2,j}^{(k+1,z)}
\end{bmatrix},
\qquad j=0,\dots,N_v-1.
\end{equation}


変数ベクトル（内部セル i=1,...,N_x-2 分）

\begin{equation}
  \mathbf{f}_{j}^{\,k+1,z}
  =
  \begin{bmatrix}
    f_{1,j}^{\,k+1,z} \\
    f_{2,j}^{\,k+1,z} \\
    \vdots \\
    f_{N_x-2,j}^{\,k+1,z}
  \end{bmatrix},
  \qquad
  j = 0,\dots,N_v-1 .
\end{equation}

ソース（右辺）ベクトル

\begin{equation}
  \mathbf{d}_{j}^{\,k+1,z-1}
  =
  \begin{bmatrix}
    d_{1,j}^{\,k+1,z-1} \\
    d_{2,j}^{\,k+1,z-1} \\
    \vdots \\
    d_{N_x-2,j}^{\,k+1,z-1}
  \end{bmatrix},
  \qquad
  j = 0,\dots,N_v-1 ,
\end{equation}

\begin{equation}
  d_{i,j}^{\,k+1,z-1}
  =
  \mathrm{EXP}_{i,j}
  + \theta^{HO,k}_{z}\,
    \frac{\Delta t}{\tau_{i}^{\,LO,k+1,z-1}}\,
    f_{M,i,j}^{\,LO,k+1,z-1},
  \qquad
  i = 1,\dots,N_x-2,\ j = 0,\dots,N_v-1 .
\end{equation}


### 差分式(LO)

k: 時間発展
z: HOLO反復
m: LO反復

\begin{align}
\frac{n_i^{\,\mathrm{LO},k+1,z}-n_i^{\,\mathrm{HO},k}}{\Delta t}
&\;+\;
\frac12\left[
  \frac{\widehat{(nu)}_{i+\frac12}^{\,\mathrm{LO},k+1,z}
        -\widehat{(nu)}_{i-\frac12}^{\,\mathrm{LO},k+1,z}}{\Delta x}
 +
  \frac{\hat S_{1,i+\frac12}^{\,\mathrm{HO},k}
        -\hat S_{1,i-\frac12}^{\,\mathrm{HO},k}}{\Delta x}
\right]
= 0 ,
\\[6pt]

\frac{(nu)_i^{\,\mathrm{LO},k+1,z}-(nu)_i^{\,\mathrm{HO},k}}{\Delta t}
&\;+\;
\frac12\left[
  \frac{\widehat{\bigl(nu^2/n+\hat P\bigr)}_{i+\frac12}^{\,\mathrm{LO},k+1,z}
        -\widehat{\bigl(nu^2/n+\hat P\bigr)}_{i-\frac12}^{\,\mathrm{LO},k+1,z}}{\Delta x}
 +
  \frac{\hat S_{2,i+\frac12}^{\,\mathrm{HO},k}
        -\hat S_{2,i-\frac12}^{\,\mathrm{HO},k}}{\Delta x}
\right]
= 0 ,
\\[6pt]

\frac{U_i^{\,\mathrm{LO},k+1,z}-U_i^{\,\mathrm{HO},k}}{\Delta t}
&\;+\;
\frac12\left[
  \frac{\widehat{\,u(U+\hat P)\,}_{i+\frac12}^{\,\mathrm{LO},k+1,z}
        -\widehat{\,u(U+\hat P)\,}_{i-\frac12}^{\,\mathrm{LO},k+1,z}}{\Delta x}
 +
  \frac{\hat S_{3,i+\frac12}^{\,\mathrm{HO},k}
        -\hat S_{3,i-\frac12}^{\,\mathrm{HO},k}}{\Delta x}
\right]
+
\frac{\widehat Q_{i+\frac12}^{\,\mathrm{HO},k+1,z-1}
      -\widehat Q_{i-\frac12}^{\,\mathrm{HO},k+1,z-1}}{\Delta x}
= 0 ,
\end{align}

\begin{equation}
\qquad i=1,\dots,N_x-2 .
\end{equation}



高次モーメント（HO, time level k, 界面 i+1/2）

### 簡易HOLO(LO系の非線形要素のみを凍結(前反復のWを使用)してフラックスベクトルを線形にするPicardで)

### picard反復index: m

モーメントベクトルWと界面平均

\begin{equation}
  \mathbf W_i^{LO,k+1,m}
  =
  \begin{bmatrix}
    n_i^{LO,k+1,m}\\[2pt]
    (nu)_i^{LO,k+1,m}\\[2pt]
    U_i^{LO,k+1,m}
  \end{bmatrix},
  \qquad m = 0,1,2,\dots
\end{equation}

\begin{equation}
  \widehat{\mathbf W}_{i+\frac12}^{\,LO,k+1,m}
  = \frac12\Bigl(
      \mathbf W_i^{LO,k+1,m}
     +\mathbf W_{i+1}^{LO,k+1,m}
    \Bigr),
  \qquad
  \widehat{\mathbf W}_{i-\frac12}^{\,LO,k+1,m}
  = \frac12\Bigl(
      \mathbf W_{i-1}^{LO,k+1,m}
     +\mathbf W_{i}^{LO,k+1,m}
    \Bigr).
\end{equation}


フラックスベクトルF

\begin{equation}
  \mathbf F(\mathbf W)
  =
  \begin{bmatrix}
    F_1(\mathbf W)\\[2pt]
    F_2(\mathbf W)\\[2pt]
    F_3(\mathbf W)
  \end{bmatrix}
  =
  \begin{bmatrix}
    (nu)\\[4pt]
    \displaystyle \frac{(nu)^2}{n} + P(\mathbf W)\\[8pt]
    u(\mathbf W)\,\bigl(U + P(\mathbf W)\bigr)
  \end{bmatrix},
  \qquad
  \mathbf W =
  \begin{bmatrix}
    n\\[2pt]
    (nu)\\[2pt]
    U
  \end{bmatrix},
\end{equation}


フラックスベクトルに含まれる非線形な要素

\begin{equation}
  u(\mathbf W) = \frac{(nu)}{n},\qquad
  T(\mathbf W) = T\bigl(n,(nu),U\bigr),\qquad
  P(\mathbf W) = n\,T(\mathbf W).
\end{equation}


非線形要素をPicard反復の前状態から事前計算

\begin{align}
  u_i^{*,m-1}
  &= u\!\left(\mathbf W_i^{LO,k+1,m-1}\right),
  &
  T_i^{*,m-1}
  &= T\!\left(\mathbf W_i^{LO,k+1,m-1}\right),
\\
  P_i^{*,m-1}
  &= n_i^{LO,k+1,m-1}\,T_i^{*,m-1},
  &&
\end{align}


\begin{align}
  u_{i+\frac12}^{*,m-1}
  &= u\!\left(
       \widehat{\mathbf W}_{i+\frac12}^{\,LO,k+1,m-1}
     \right),
\\
  T_{i+\frac12}^{*,m-1}
  &= T\!\left(
       \widehat{\mathbf W}_{i+\frac12}^{\,LO,k+1,m-1}
     \right),
\\
  P_{i+\frac12}^{*,m-1}
  &= \widehat n_{i+\frac12}^{\,LO,k+1,m-1}\,
     T_{i+\frac12}^{*,m-1}
\end{align}


非線形要素を凍結しPicardにより線形化したフラックスベクトル

\begin{equation}
  \widetilde{\mathbf F}^{(m)}(\mathbf W_i)
  :=
  \begin{bmatrix}
    (nu)_i \\[4pt]
    \bigl(u_i^{*,m-1}\bigr)^2\,n_i + P_i^{*,m-1}\\[6pt]
    u_i^{*,m-1}\,U_i + u_i^{*,m-1}\,P_i^{*,m-1}
  \end{bmatrix}.
\end{equation}


\begin{equation}
  \widetilde{\mathbf F}^{(m)}(\mathbf W_i)
  = A_i^{(m-1)}\,\mathbf W_i + \mathbf b_i^{(m-1)}
\end{equation}


\begin{equation}
  A_i^{(m-1)} =
  \begin{bmatrix}
    0 & 1 & 0\\[2pt]
    \bigl(u_i^{*,m-1}\bigr)^2 & 0 & 0\\[2pt]
    0 & 0 & u_i^{*,m-1}
  \end{bmatrix},
  \qquad
  \mathbf b_i^{(m-1)} =
  \begin{bmatrix}
    0\\[2pt]
    P_i^{*,m-1}\\[2pt]
    u_i^{*,m-1}\,P_i^{*,m-1}
  \end{bmatrix}.
\end{equation}


フラックスベクトルの書き下し

\begin{equation}
  \widetilde{\mathbf F}^{(m)}_{i+\frac12}
  :=
  \widetilde{\mathbf F}^{(m)}\!\left(
    \widehat{\mathbf W}_{i+\frac12}^{\,LO,k+1,m}
