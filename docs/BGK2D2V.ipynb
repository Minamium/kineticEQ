## BBGK 2D2Vの連続式

\begin{equation}
  \partial_t f(t,x,y,v_x,v_y)
  + v_x \,\partial_x f(t,x,y,v_x,v_y)
  + v_y \,\partial_y f(t,x,y,v_x,v_y)
  = \frac{M[f](t,x,y,v_x,v_y) - f(t,x,y,v_x,v_y)}{\tau(n(t,x,y),T(t,x,y))}.
\end{equation}

### 0th moment

\begin{equation}
  n(t,x,y)
  = \int_{\mathbb{R}^2} f(t,x,y,v_x,v_y)\, dv_x dv_y.
\end{equation}

### 1st moment

\begin{equation}
  \begin{aligned}
    n(t,x,y) u_x(t,x,y)
    &= \int_{\mathbb{R}^2} v_x \, f(t,x,y,v_x,v_y)\, dv_x dv_y,\\
    n(t,x,y) u_y(t,x,y)
    &= \int_{\mathbb{R}^2} v_y \, f(t,x,y,v_x,v_y)\, dv_x dv_y.
  \end{aligned}
\end{equation}

### 2nd moment

\begin{equation}
  T(t,x,y)
  = \frac{1}{2\,n(t,x,y)}
    \int_{\mathbb{R}^2}
      \bigl[(v_x - u_x(t,x,y))^2 + (v_y - u_y(t,x,y))^2\bigr]
      f(t,x,y,v_x,v_y)\, dv_x dv_y.
\end{equation}

### マクスウェル分布

\begin{equation}
  M[f](t,x,y,v_x,v_y)
  =
  \frac{n(t,x,y)}{2\pi\,T(t,x,y)}
  \exp\left(
    -\frac{
      (v_x - u_x(t,x,y))^2
      + (v_y - u_y(t,x,y))^2
    }{
      2\,T(t,x,y)
    }
  \right).
\end{equation}

### 緩和時間τ

\begin{equation}
  \tau\bigl(n(t,x,y),T(t,x,y)\bigr)
  =
  \frac{\tilde{\tau}}{
    n(t,x,y)\,\sqrt{T(t,x,y)}
  }.
\end{equation}


## Discretazation

\begin{equation}
  x_i = x_{\min} + i\,\Delta x,\quad i = 0,1,\dots,N_x,
  \quad
  y_j = y_{\min} + j\,\Delta y,\quad j = 0,1,\dots,N_y,
  \quad
  v_{x,p} = v_{x,\min} + p\,\Delta v_x,\quad p = 0,1,\dots,N_{v_x},
  \quad
  v_{y,q} = v_{y,\min} + q\,\Delta v_y,\quad q = 0,1,\dots,N_{v_y}.
\end{equation}

\begin{equation}
  t^k = k\,\Delta t,\quad k = 0,1,2,\dots
\end{equation}

\begin{equation}
  f_{i,j,p,q}^k \approx f\bigl(t^k, x_i, y_j, v_{x,p}, v_{y,q}\bigr).
\end{equation}


### moments discretazation

\begin{equation}
  n_{i,j}^k
  =
  \sum_{p=0}^{N_{v_x}}
  \sum_{q=0}^{N_{v_y}}
    f_{i,j,p,q}^k\,
    \Delta v_x\,\Delta v_y.
\end{equation}

\begin{equation}
  n_{i,j}^k\,u_{x,i,j}^k
  =
  \sum_{p=0}^{N_{v_x}}
  \sum_{q=0}^{N_{v_y}}
    v_{x,p}\,f_{i,j,p,q}^k\,
    \Delta v_x\,\Delta v_y,
\end{equation}

\begin{equation}
  n_{i,j}^k\,u_{y,i,j}^k
  =
  \sum_{p=0}^{N_{v_x}}
  \sum_{q=0}^{N_{v_y}}
    v_{y,q}\,f_{i,j,p,q}^k\,
    \Delta v_x\,\Delta v_y.
\end{equation}

\begin{equation}
  T_{i,j}^k
  =
  \frac{1}{2\,n_{i,j}^k}
  \sum_{p=0}^{N_{v_x}}
  \sum_{q=0}^{N_{v_y}}
    \Bigl[
      (v_{x,p} - u_{x,i,j}^k)^2
      +
      (v_{y,q} - u_{y,i,j}^k)^2
    \Bigr]
    f_{i,j,p,q}^k\,
    \Delta v_x\,\Delta v_y.
\end{equation}



### Explicit 

\begin{equation}
  D_x f_{i,j,p,q}^k
  =
  \begin{cases}
    \displaystyle
    \dfrac{f_{i,j,p,q}^k - f_{i-1,j,p,q}^k}{\Delta x},
      & v_{x,p} > 0,\\[1.2ex]
    \displaystyle
    \dfrac{f_{i+1,j,p,q}^k - f_{i,j,p,q}^k}{\Delta x},
      & v_{x,p} < 0,\\[1.2ex]
    0, & v_{x,p} = 0.
  \end{cases}
\end{equation}

\begin{equation}
  D_y f_{i,j,p,q}^k
  =
  \begin{cases}
    \displaystyle
    \dfrac{f_{i,j,p,q}^k - f_{i,j-1,p,q}^k}{\Delta y},
      & v_{y,q} > 0,\\[1.2ex]
    \displaystyle
    \dfrac{f_{i,j+1,p,q}^k - f_{i,j,p,q}^k}{\Delta y},
      & v_{y,q} < 0,\\[1.2ex]
    0, & v_{y,q} = 0.
  \end{cases}
\end{equation}


### Maxwellian distribution

\begin{equation}
  M_{i,j,p,q}^k
  =
  \frac{n_{i,j}^k}{2\pi\,T_{i,j}^k}
  \exp\left(
    -\frac{
      (v_{x,p} - u_{x,i,j}^k)^2
      + (v_{y,q} - u_{y,i,j}^k)^2
    }{
      2\,T_{i,j}^k
    }
  \right),
\end{equation}

\begin{equation}
  \tau_{i,j}^k
  =
  \frac{\tilde{\tau}}{
    n_{i,j}^k\,\sqrt{T_{i,j}^k}
  }.
\end{equation}


### Explicit Discretization BBGK equation

The explicit discretization of the BBGK equation is given by:

\begin{equation}
  \begin{aligned}
    f_{i,j,p,q}^{k+1}
    &=
    f_{i,j,p,q}^{k}
    - \Delta t\,
      \Bigl(
        v_{x,p}\,D_x f_{i,j,p,q}^{k}
        +
        v_{y,q}\,D_y f_{i,j,p,q}^{k}
      \Bigr)
    \\
    &\qquad
    + \Delta t\,
      \frac{
        M_{i,j,p,q}^{k}
        - f_{i,j,p,q}^{k}
      }{
        \tau_{i,j}^{k}
      }.
  \end{aligned}
\end{equation}


\begin{equation}
  v_{x,\max}
  = \max_{p} \bigl| v_{x,p} \bigr|,
  \qquad
  v_{y,\max}
  = \max_{q} \bigl| v_{y,q} \bigr|.
\end{equation}


\begin{equation}
  \Delta t
  \le
  \mathrm{CFL}_{\mathrm{adv}}
  \,\min\!\left(
    \frac{\Delta x}{v_{x,\max}},
    \frac{\Delta y}{v_{y,\max}}
  \right),
  \qquad
  0 < \mathrm{CFL}_{\mathrm{adv}} \le 1.
\end{equation}


\begin{equation}
  \tau_{\min}
  =
  \min_{i,j,k}
  \tau_{i,j}^k
  =
  \min_{i,j,k}
  \frac{\tilde{\tau}}{
    n_{i,j}^k \sqrt{T_{i,j}^k}
  }.
\end{equation}

\begin{equation}
  \Delta t
  \le
  \mathrm{CFL}_{\mathrm{col}}\,
  \tau_{\min},
  \qquad
  0 < \mathrm{CFL}_{\mathrm{col}} \le 1.
\end{equation}


\begin{equation}
  \Delta t
  \le
  \min\!\left[
    \mathrm{CFL}_{\mathrm{adv}}
    \,\min\!\left(
      \frac{\Delta x}{v_{x,\max}},
      \frac{\Delta y}{v_{y,\max}}
    \right),
    \;
    \mathrm{CFL}_{\mathrm{col}}\,
    \tau_{\min}
  \right].
\end{equation}
